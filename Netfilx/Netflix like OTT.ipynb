```python
import random
import re
import mysql.connector


db = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="********",
    database="movies"
)

cursor = db.cursor()



try:
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS movies (
            id INT AUTO_INCREMENT PRIMARY KEY,
            title VARCHAR(255) NOT NULL,
            year INT NOT NULL,
            genre VARCHAR(50),
            rating FLOAT
        )
    ''')
except:
    pass # Ignore exception if table already exists

 


sample_movies = [
    (1, 'Inception', 2010, 'Sci-Fi', 8.8),
    (2, 'The Dark Knight', 2008, 'Action', 9.0),
    (3, 'Pulp Fiction', 1994, 'Drama', 8.9),
    (4, 'The Shawshank Redemption', 1994, 'Drama', 9.3),
    (5, "Drak", 2013, "Comedy", 8.0),
    (6, "Bright", 2018, "Drama", 9.8)
]

for movie in sample_movies:
    cursor.execute('INSERT INTO movies (id, title, year, genre, rating) VALUES (%s, %s, %s, %s, %s)', movie)


db.commit()

######################
# Function to get movie recommendations
def get_movie_recommendation(genre=None, rating=None):
    # Build SQL query
    query = "SELECT * FROM movies"

    if genre:
        query += f" WHERE genre = '{genre}'"

    if rating:
        if genre:
            query += f" AND rating >= {rating}"
        else:
            query += f" WHERE rating >= {rating}"

    cursor.execute(query)
    movies = cursor.fetchall()

    if not movies:
        return "No movies found with the given criteria."

    return random.choice(movies)

# Lambda function for formatting movie details
format_movie = lambda movie: f"{movie[1]} ({movie[2]}) - Genre: {movie[3]}, Rating: {movie[4]}"

# List of genres
genres = ["Action", "Comedy", "Drama", "Sci-Fi", "Horror"]

# Get user input for minimum rating
user_rating_input = input("Enter the minimum rating you want for movie recommendations: ")
try:
    user_rating = float(user_rating_input)
except ValueError:
    print("Invalid input. Please enter a valid number.")
    exit()

# Recommend at least one movie
recommended_movie = get_movie_recommendation(genre=random.choice(genres), rating=user_rating)
print("Recommended Movie:", format_movie(recommended_movie))


############




# Tuple 
example_movie = (1, "Inception", 2010, "Sci-Fi", 8.8)


# List comprehension 
recommendations = [get_movie_recommendation(genre=random.choice(genres), rating=random.uniform(7.0, 9.0)) for _ in range(5)]


class RecommendationError(Exception):
    def __init__(self, message="Error in movie recommendation"):
        self.message = message
        super().__init__(self.message)


try:
    raise RecommendationError("No recommendations found.")
except RecommendationError as e:
    print(f"Exception: {e}")


iterator = iter(recommendations)
for i in range(len(recommendations)):
    print(f"Recommendation {i+1}: {next(iterator)}")


title = "Star Wars Episode IV: A New Hope (1977)"
year = re.search(r'\((\d{4})\)', title).group(1)
print(f"Movie title: {title}, Year: {year}")


cursor.close()
db.close()
```
